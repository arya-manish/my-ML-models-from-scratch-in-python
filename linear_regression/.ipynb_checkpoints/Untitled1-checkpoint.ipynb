```python
# Sigmoid (or Logistic) Function -  #Helper function(we use it in forward_prop)

def sigmoid(z):  #z is a scalar or numpy array 
    a = 1/(1+np.exp(-z)) 
    return a
```

```python
#step1
def initialize_parameters(lenw):
    w = np.random.randn(1,lenw)
    # w = np.zeros((1,lenw))
    b = 0
    return w,b
```

```python
#step2
def forward_prop(X,w,b):  #w-->1xn , X-->nxm
    z = np.dot(w,X)+ b  #z--> 1xm  b_vector = [b b b  ....]
    a = sigmoid(z)
    return a
```

```python
#step3
def cost_function(y,a):
    m = y.shape[1]
    L = np.multiply(y,np.log(a)+np.multiply(1-y,np.log(1-a)))
    J = -(1/m)*np.sum(L)
    return J
```

```python
#step4
def back_prop(X,y,a):
    m = y.shape[1]
    dz = (1/m)*(a-y)
    dw = np.dot(dz,X.T)
    db = np.sum(dz)
    return dw, db
```

```python
#step5
def gradient_descent_update(w,b,dw,db,learning_rate):
    w = w - learning_rate*dw
    b = b - learning_rate*db
    return w,b
```

```python
def accuracy_of_prediction(y,y_prediction):
    m = y.shape[1]
    accuracy_vector = y_prediction==y
    #print(accuracy_vector)
    accuracy = np.sum(accuracy_vector)/m
    return accuracy
```

```python
def logistic_regression_model(X_train, y_train, X_val, y_val, learning_rate, epochs):
    
    lenw = X_train.shape[0]
    w,b = initialize_parameters(lenw)  #step1
    
    costs_train = []
    m_train = y_train.shape[1]
    m_val = y_val.shape[1]
    
    for i in range(1,epochs+1):
        a_train = forward_prop(X_train,w,b) #step2
        cost_train = cost_function(y_train,a_train,) #step3
        dw,db = back_prop(X_train,y_train,a_train)  #step4
        w,b = gradient_descent_update(w,b,dw,db,learning_rate) #step5
        
         # store training cost in a list for plotting purpose
        if i%10==0:
            costs_train.append(cost_train)
            
         # accuracy of prediction on training set
        y_prediction_train = prediction(y_train,a_train)
        accuracy_train = accuracy_of_prediction(y_train,y_prediction_train)
        
        # cost & accuracy of prediction on validation set
        a_val = forward_prop(X_val,w,b)
        cost_val = cost_function(y_val,a_val)
        y_prediction_val = prediction(y_val,a_val)
        accuracy_val = accuracy_of_prediction(y_val,y_prediction_val)
        
        # print out cost_train,cost_val,accuracy_train, accuracy_val
        
        print('Epochs '+str(i)+'/'+str(epochs)+': ')
        print('Training cost '+str(cost_train)+'|'+'Validation cost '+str(cost_val))
        print('Training accuracy '+str(accuracy_train)+'|'+'Validation accuracy '+str(accuracy_val))
        
    # plotting training cost as a function of iterations(per tens)
    plt.plot(costs_train)
    plt.xlabel('Iterations(per tens)')
    plt.ylabel('Training cost')
    plt.title('Learning rate '+str(learning_rate))
    plt.show()
    
    #return w,b,accuracy_val
        
        
        
```
